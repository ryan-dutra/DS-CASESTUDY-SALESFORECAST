```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_absolute_error
```

```python
train, test = pd.read_parquet('../data/train.snap.parquet'), pd.read_parquet('../data/test.snap.parquet')
```

```python
train['weekday'] = train['date'].dt.weekday
weekly_average = train.groupby(['item_id', 'store_id', 'weekday'])['y'].mean().reset_index()
weekly_average.rename(columns={'y': 'y_weekday_mean'}, inplace=True)
```

```python
weekly_average
```

```python
test['weekday'] = test['date'].dt.weekday
test = test[['item_id', 'store_id', 'weekday', 'd', 'y']]
test
```

```python
test = test.merge(weekly_average[['item_id', 'store_id', 'weekday', 'y_weekday_mean']], 
                  on=['item_id', 'store_id', 'weekday'], 
                  how='left')
test
```

```python
mae = mean_absolute_error(test['y'], test['y_weekday_mean'])
print(f'Mean Absolute Error (MAE): {mae}')
```

```python
days = [f'd_{c}' for c in range(1942, 1970)]
for d in days:
    scope = test.copy()
    scope = scope[scope['d'] == d]
    mae = mean_absolute_error(scope['y'], scope['y_weekday_mean'])
    print(f'Mean Absolute Error (MAE) for {d}: {mae}')
    
    
```
